```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Pi Estimation

```python
def is_within(p):
    dist = np.sqrt(np.sum(p ** 2))
    return dist <= 1

points = np.random.rand(2, 5000) * 2 - 1
labels = np.apply_along_axis(is_within, 0, points)
    
plt.subplots(1, 1, figsize=(7, 7))
plt.scatter(points[0, :], points[1, :], c=['r' if l else 'b' for l in labels])
plt.show()
```

Red points lie inside a unit circle $c$ of area $A_c = \pi r^2$
where $r=1$. All of the points lie within a square $s$ with area
$A_s = 4r^2$. The Ratio $R$ of the areas is

$$R = \frac{A_c}{A_s} = \frac{\pi r^2}{4 r^2}= \frac{\pi}{4}$$

which can be aproximated as

$$R \approx \frac{N_c}{N_t}$$

where $N_c$ is the number of points lying within the unit circle and $N_s$ is the
total number of points. Value of $\pi$ can then be estimated as

$$\pi \approx 4\frac{N_c}{N_t}$$

```python
c_s_ratio = labels.mean()
pi_est = 4 * c_s_ratio
print(f'Pi = {pi_est:.4f}')
```
