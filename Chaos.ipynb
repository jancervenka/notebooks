```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
class Chaos:
    
    def __init__(self, x0, r):
        self.r = r
        self.x = x0
        
    def update(self):
        self.x = self.r * self.x * (1 - self.x)
        
    def generate_sequence(self, steps):
        path = []
        for i in range(steps):
            path.append(self.x)
            self.update()
    
        return path
```

```python
ch = Chaos(0.1, 3.5)
out = ch.generate_sequence(75)
plt.subplots(1, 1, figsize=(8, 6))
plt.plot(out, 'b')
plt.xlabel('time [-]')
plt.ylabel('val [-]')
plt.title('Logistic Map Example, r = ' + str(ch.r))
plt.show()
```

```python
ch = Chaos(0.1, 2.5)
fig, ax = plt.subplots(figsize=(8, 6))
while ch.r < 4:
    ch.r += 0.005
    seq = ch.generate_sequence(1000)[950: ]
    ax.plot(np.zeros(len(seq)) + ch.r, seq, 'b.', ms=0.6)

ax.set_xlabel('r [-]')
ax.set_ylabel('val [-]')
ax.set_title('Logistic Map (Bifurcation Diagram)')
plt.show()
```
