```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Chaos:
    
    def __init__(self, x0, r):
        self.r = r
        self.x = x0
        
    def set_r(self, r):
        self.r = r
        
    def get_r(self):
        return self.r
        
    def update(self):
        self.x = self.r * self.x * (1 - self.x)
        
    def generate_sequence(self, steps):
        path = []
        for i in range(steps):
            path.append(self.x)
            self.update()
    
        return path
```

```python
ch = Chaos(0.1, 3.5)
out = ch.generate_sequence(75)
plt.plot(out)
plt.xlabel('time [-]')
plt.ylabel('val [-]')
plt.title('Logistic Map Example, r = ' + str(ch.get_r()))
plt.show()
```

```python
r = 2.5
ch = Chaos(0.1, r)
fig, ax = plt.subplots()
while r < 4:
    ch.set_r(r)
    r += 0.005
    seq = ch.generate_sequence(1000)[950 : ]
    ax.plot(np.zeros(len(seq)) + r, seq, 'b.', ms=  0.6)

ax.set_xlabel('r [-]')
ax.set_ylabel('val [-]')
ax.set_title('Logistic Map (Bifurcation Diagram)')
plt.show()
```
