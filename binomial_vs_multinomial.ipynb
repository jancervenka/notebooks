```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# coin toss (p=0.5), binomial (n=20 attempts)
# https://en.wikipedia.org/wiki/Binomial_distribution

x = np.random.binomial(n=20, p=0.5, size=1000)
plt.hist(x, bins=20)
plt.show()
```

```python
# coin toss p=0.5, berounulli (n=1 attempts)
x = np.random.binomial(n=1, p=0.5, size=1000)
plt.hist(x)
plt.show()
```

```python
# dice p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6], multinomial (n=30 attempts)
# https://en.wikipedia.org/wiki/Multinomial_distribution
# For example, it models the probability of counts for each side of a k-sided die rolled n times
# f(x1, x2,... xk; p1, p2,... pk, n)

X = np.random.multinomial(n=30, pvals=[1 / 6] * 6, size=1000)
print(X.shape)

fig, ax = plt.subplots(2, 3, figsize=(10, 7))
ax = ax.flatten()

# each side ladning 5 times (30 / 6) is the most common outcome
for i in range(6):
    ax[i].hist(X[:, i])
    ax[i].set_title(f'Side={i}')

plt.show()
```

```python
# dice p = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6], multinoulli (n=1 attempts)
X = np.random.multinomial(n=1, pvals=[1 / 6] * 6, size=1000)
print(X.shape)

fig, ax = plt.subplots(2, 3, figsize=(10, 7))
ax = ax.flatten()

# each side ladning 0 times out of one attempt is the most common outcome
for i in range(6):
    ax[i].hist(X[:, i])
    ax[i].set_title(f'Side={i}')

plt.show()
```
