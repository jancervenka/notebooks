```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

PDF $f(x)$ of $\mathcal{N}(0, 1)$ is

$$
f(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2}
$$

$$
\frac{\mathrm{d}f(x)}{\mathrm{d}x} =
-x\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2}
$$

$$
\frac{\mathrm{d}^2f(x)}{\mathrm{d}x^2} =
x^2\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2} -
\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2} =
\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2}(x^2 - 1)
$$

```python
def f(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x ** 2)

def d2f(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x ** 2) * (x ** 2 -1)
```

```python
x = np.arange(-5, 5, 0.01)

plt.subplots(1, 1, figsize=(7, 5))
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot([-1, -1], [-1, 1], 'r--')
plt.plot([1, 1], [-1, 1], 'r--')

plt.plot(x, d2f(x), 'g', label='d2f(x)')

plt.legend()
plt.title('N(0, 1) Inflection Points')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0, 0.5])
plt.show()
```
